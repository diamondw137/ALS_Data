```python
subs = pd.read_csv('cons_email_chapter_subscription.csv')
subs.head()
```

```python
subs.count()
```

```python
subs = subs.rename(columns={'cons_email_id':'cons_id'})
subs.head()
                   
```

```python
merge1 = pd.merge(email, subs, on='cons_id', how='left')
merge1.head()
```

```python
merge1.count()
```

```python
people = merge1[['email','cons_id','note','isunsub','create_dt','modified_dt_x']]
people.head()
```

```python
people.to_csv('people.csv')
```

```python
acquisitions = pd.to_datetime(people['create_dt'])
acquisitions.head()
```

```python
acquisitions = acquisitions.groupby(acquisitions.dt.date).size().reset_index(name='Count')
acquisitions.head()
```

```python
acquisitions.to_csv('acquisitions.csv')
```
